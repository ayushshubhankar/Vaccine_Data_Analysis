# Plotting and Visualisation



```python
#setting figure size to accomodate all the countries
fig,ax=plt.subplots(figsize=(50,90))
#setting font size for better viewing
plt.rcParams.update({'font.size':100})
# Plotting a horizontal bar graph
ax.barh(df["Country"],df["Doses administered per 100 people"])
#setting relevent title and labels
ax.set(title="Doses administered per 100 people by country",
      xlabel="Number of Doses",ylabel="Countries")
# Creating a mean line to show the mean
ax.axvline(df["Doses administered per 100 people"].mean(),linestyle="--",color="red")
```

 * We see the plot of various countries and their number of doses per 100 of their population

```python
fig,ax=plt.subplots()
ax.scatter(df["Doses administered per 100 people"],df["% of population fully vaccinated"])
ax.set(title="relation between doses and percentage fully vaccinated",xlabel="Doses administered per 100 people",ylabel="% of population fully vaccinated")
```

#  Inference from both the plots
* As the number of doses administered increases, percentage of population vaccinated also increases.
* some countries have vaccinated people less than the average of the total population,whereas some more than the mean.
